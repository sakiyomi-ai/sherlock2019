# 1. replace ‘_’ and ’T’ by ‘ ‘

```python
import os

inList = ["ACaseOfIdentity.txt", "SpeckledBand.txt", "CrookedMan.txt", "DevilsFoot.txt", "DancingMen.txt"]
outList = ["ACaseOfIdentity_formatted.txt", "SpeckledBand_formatted.txt", "CrookedMan_formatted.txt", "DevilsFoot_formatted.txt", "DancingMen_formatted.txt"]

listLen = len(inList)

for index in range(listLen):
    fileIn = open(inList[index], 'r')
    fileOut = open(outList[index], 'w')

    for line in fileIn:
        tokens = line.rstrip('\n').split(",")
        s0 = tokens[0]
        
        if (tokens[1].startswith("_") is False) and (tokens[1].endswith("_") is False):
            s1 = tokens[1].replace('_', ' ')
        else:
            s1 = tokens[1]
    
        if (tokens[2].startswith("_") is False) and (tokens[2].endswith("_") is False):
            s2 = tokens[2].replace('_', ' ')
        else:
            s2 = tokens[2]

        if(s1 == 'time'):
            s2 = s2.replace('T', ' ')

        fileOut.write(s0 + "," + s1 + "," + s2 + "\n" )

    fileIn.close()
    fileOut.close()
```

# 2. create formatted file

```python
import os

inList = ["ACaseOfIdentity_formatted.txt", "SpeckledBand_formatted.txt", "CrookedMan_formatted.txt", "DevilsFoot_formatted.txt", "DancingMen_formatted.txt"]
outList = ["ACaseOfIdentity_formatted_S2.txt", "SpeckledBand_formatted_S2.txt", "CrookedMan_formatted_S2.txt", "DevilsFoot_formatted_S2.txt", "DancingMen_formatted_S2.txt"]
errorList = ["ACaseOfIdentity-error.txt", "SpeckledBand-error.txt", "CrookedMan-error.txt", "DevilsFoot-error.txt", "DancingMen-error.txt"]

def findStrange(line, fileError):
    result = ""
    s1 = line.rstrip('\n')
    parts = s1.split(',')

    for part in parts:
        sub_parts = part.split(' ')

        if part == 'metadata':
            for part in parts:
                fileError.write(part + ',')       

            fileError.write('\n')
            result = ''
            return(result)

        if part.isdigit() is True:
            result = result + part + ','
        else:
            listlen = len(sub_parts)
            
            for sub_part in sub_parts:
                if(sub_part ==''):
                    for part in parts:
                        fileError.write(part + ',')       

                    fileError.write('\n')
                    result = ''
                    return(result)
            
                a = sub_part[0]
                if sub_part.islower() is False and a.islower() is True:
                    for w in sub_part:
                        if w.isupper() is True:
                            index = sub_part.find(w)
                            s2 = sub_part.replace(w, ' ' + w) 
                            sub_part = s2
                        else:
                            pass
                        if listlen >1:
                            result = result + sub_part + ' '
                        else:
                            result = result + sub_part + ','                        
                else:
                    if listlen >1:
                        result = result + sub_part + ' '
                    else:
                        result = result + sub_part + ','

                strLen = len(result)
                if(strLen > 0):
                    if result[strLen -1] != ',' and result[strLen -1] == ' ' :
                        s3 = result.strip(' ')   
                        result = s3 + ','
                    else:
                        pass
    
    # 変換完了
    strLen = len(result)
    if(strLen > 0):
        if result[strLen -1] == ',':
            s4 = result.strip(',')        
            result = s4
        else:
            pass

        result = result + '\n'
    else:
        result = ""
    #print(result)

    return result
```

```python
listLen = len(inList)
for index in range(listLen):
    fileIn = open(inList[index], 'r')
    fileOut = open(outList[index], 'w')
    fileError = open(errorList[index], 'w')

    for line in fileIn:
        result = findStrange(line, fileError)

        if result != 0:
            fileOut.write(result)

    fileIn.close()
    fileOut.close()
    fileError.close()
```
